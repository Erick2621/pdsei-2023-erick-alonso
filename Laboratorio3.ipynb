Implementa el algoritmo de especificaci√≥n de histograma de una imagen en escala de grises, cuyo procedimiento se resume en la figura 3

```python
IMAGEN_BASE_MACHU_PICHU = "../content/drive/MyDrive/LAB3 PDSEI/machu_pichu.jpg"
IMAGEN_MODIFICAR_SACSAYHUAMAN = "/content/drive/MyDrive/LAB3 PDSEI/Sacsayhuaman.png"
```

```python
#SE LEE LA IMAGEN DE BASE Y SE PASA A ESCALA DE GRISES
IMAGEN_BASE = cv2.imread(IMAGEN_BASE_MACHU_PICHU, 0)
IMAGEN_BASE_GRAY = cv2.cvtColor(IMAGEN_BASE, cv2.COLOR_BGR2RGB)
np.histogram (IMAGEN_BASE_GRAY, 256, [0,256])
```

```python
plt.imshow(IMAGEN_BASE_GRAY)
```

```python
#HISTOGRAMA DE LA IMAGEN BASE
hist, bins = np.histogram (IMAGEN_BASE_GRAY, 256, [0,256])
plt.bar(np.arange(256), hist)
plt.show()
```

```python
color = ('r*', 'gs', 'b@')
for i, col in enumerate(color):
    hist=cv2.calcHist([IMAGEN_BASE_GRAY], [0], None,[256], [0,256])
#       LA IMAGEN SE TOMA DE LA IMAGEN LEIDA
    plt.plot(hist)
```

```python
#SE LEE LA IMAGEN A MODIFICAR Y SE PASA A ESCALA DE GRISES
IMAGEN_MOD = cv2.imread(IMAGEN_MODIFICAR_SACSAYHUAMAN,0)
IMAGEN_MOD_GRAY = cv2.cvtColor(IMAGEN_MOD, cv2.COLOR_BGR2RGB)
np.histogram (IMAGEN_MOD_GRAY, 256, [0,256])
```

```python
plt.imshow(IMAGEN_MOD_GRAY)
```

```python
#HISTOGRAMA DE LA IMAGEN A MODIFICAR
hist1, bins1 = np.histogram (IMAGEN_MOD_GRAY, 256, [0,256])
plt.bar(np.arange(256), hist1)
plt.show()
```

```python
color = ('r*', 'gs', 'b@')
for i, col in enumerate(color):
    hist=cv2.calcHist([IMAGEN_MOD_GRAY], [0], None,[256], [0,256])
#       LA IMAGEN SE TOMA DEL PLOTEO ANTERIOR
    plt.plot(hist)
```

```python
# Convertir las im√°genes a escala de grises
imagen_a_gris = cv2.cvtColor(IMAGEN_BASE_GRAY, cv2.COLOR_BGR2GRAY)
imagen_b_gris = cv2.cvtColor(IMAGEN_MOD_GRAY, cv2.COLOR_BGR2GRAY)

# Calcular el histograma acumulativo de la imagen A
hist_a = cv2.calcHist([imagen_a_gris], [0], None, [256], [0, 256])
cdf_a = hist_a.cumsum()
cdf_a_normalized = cdf_a / cdf_a[-1]

# Calcular el histograma acumulativo de la imagen B
hist_b = cv2.calcHist([imagen_b_gris], [0], None, [256], [0, 256])
cdf_b = hist_b.cumsum()
cdf_b_normalized = cdf_b / cdf_b[-1]

# Crear una tabla de mapeo para la transformaci√≥n
lut = np.interp(cdf_b_normalized, cdf_a_normalized, np.arange(256))

# Aplicar la transformaci√≥n a la imagen B
imagen_b_ecualizada = cv2.LUT(imagen_b_gris, lut)

plt.imshow(imagen_b_ecualizada, cmap='gray')
```

```python
np.histogram (imagen_b_ecualizada, 256, [0,256])
```

##Averig√ºe como crear un histograma en 2d con OpenCV.

```python
#OpenCV
img2= cv2.imread('../Imagenes/delfin.jpg')
cv2.imshow('Image', img2)
cv2.waitKey(0)

channels = cv2.split(img2)
channel_pairs = [(channels[1], channels[0]), (channels[1], channels[2]), (channels[0], channels[2])]
pairs_titles=['G&B','G&R','B&R']

fig = plt.figure()

for position, (chs, title) in enumerate (zip (channel_pairs, pairs_titles), start=1):
    # A√±adimos un subgr√°fico correspondiente al par actual.
    ax = fig.add_subplot(130+ position)

    # Calculamos el histograma
    histogram = cv2.calcHist(chs, [0, 1], None, [32, 32], [0, 256, 0, 256])
    # Dibujamos el histograma y lo titulamos.
    p = ax.imshow(histogram, interpolation="nearest")
    ax.set_title(f'20 Hist. de {title}')

    # A√±adimos a la leyenda una barra de color para ayudarnos a interpretar el histograma.
    plt.colorbar(p)
# Mostramos el resultado.
plt.show()
#Imprimimos las dimensiones del histograma.
print(f'Histograma 20 de dimensiones {histogram.shape} con un total de {histogram.flatten().shape[0]} valores.')

# Cerramos Los gr√°ficos y ventanas creadas.
plt.close()
cv2.destroyAllWindows()
```

```python
#OpenCV (otra forma)
img2= cv2.imread('../Imagenes/delfin.jpg')
channels = cv2.split(img2)
colors = ("b", "g", "r")        # Inicializar una tupla
plt.figure()
plt.title("Color de Histograma")
plt.xlabel("Bins")
plt.ylabel("Numero de Pixeles")

for (i, col) in zip(channels, colors):       # Recorrer los canales de la imagen
    hist = cv2.calcHist([i], [0], None, [256], [0, 256])   # Crear un histograma para el canal actual
    plt.plot(hist, color = col)      # Plotea el histograma
    plt.xlim([0, 256])
```

##Defina una funci√≥n que ingresando un rango de entrada (ej. 100, 120), se cree una funci√≥n para expandir ese rango, en todo el rango din√°mico de 0 a 255.

```python
def expandir_rango():
    valor_min = 0
    valor_max = 255

    # Solicitar al usuario que ingrese los valores del rango de entrada
    rango_entrada_str = input("Ingrese el rango de entrada (por ejemplo, 100,120): ")
    rango_entrada = tuple(map(int, rango_entrada_str.split(',')))

    # Calcula la diferencia entre los l√≠mites del rango de entrada
    diferencia_entrada = rango_entrada[1] - rango_entrada[0]

    # Calcula la diferencia entre los l√≠mites del rango de salida
    diferencia_salida = valor_max - valor_min

    def expansion(valor):
        # Aplica la expansi√≥n del valor en el rango de entrada al rango de salida
        valor_expandido = valor_min + (valor - rango_entrada[0]) * (diferencia_salida / diferencia_entrada)

        # Asegura que el valor expandido est√© dentro del rango [valor_min, valor_max]
        valor_expandido = max(valor_min, min(valor_max, valor_expandido))

        return valor_expandido

    return expansion

# Ejemplo de uso:
expansion_fn = expandir_rango()

while True:
    valor_str = input("Ingrese un valor para expandir (o 'q' para salir): ")
    if valor_str.lower() == 'q':
        break
    valor = int(valor_str)
    valor_expandido = expansion_fn(valor)
    print(f"Valor expandido: {valor_expandido}")
```

##Definir una funci√≥n para aumentar o reducir el brillo de una imagen, seg√∫n valor especificado.

```python
def ajustar_brillo_color(imagen, valor_brillo):
    imagen_ajustada = np.clip(imagen + valor_brillo, 0, 255)
    return imagen_ajustada

# Carga la imagen original
img = cv2.imread('scarlett_oc.png')

# Solicita al usuario ingresar el valor para aumentar o reducir el brillo
valor_brillo = int(input("Ingresa el valor para ajustar el brillo (positivo para aumentar, negativo para reducir): "))

# Ajusta el brillo en la imagen a color
img_brillo_ajustado = ajustar_brillo_color(img, valor_brillo)

# Asegura que los valores de la imagen est√©n en el rango [0, 255]
img_brillo_ajustado = np.clip(img_brillo_ajustado, 0, 255).astype(np.uint8)

# Muestra las im√°genes original y ajustada
plt.figure(figsize=(12, 6))

plt.subplot(121)
plt.title('Original')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(122)
plt.title('Brillo Ajustado')
plt.imshow(cv2.cvtColor(img_brillo_ajustado, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.tight_layout()
plt.show()
```
