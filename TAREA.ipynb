2.	 Implemente un algoritmo que le permita realizar la operación de balance de blancos.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def balance_de_blancos_yuv(imagen):
    # Paso 1: Convertir la imagen RGB al espacio YUV
    imagen_yuv = cv2.cvtColor(imagen, cv2.COLOR_BGR2YUV)

    # Paso 2: Calcular la media de los canales U y V
    media_u = np.mean(imagen_yuv[:,:,1])
    media_v = np.mean(imagen_yuv[:,:,2])

    # Paso 3: Modificar U y V (suma) de manera que la media sea 128
    delta_u = 128 - media_u
    delta_v = 128 - media_v

    imagen_yuv[:,:,1] = np.clip(imagen_yuv[:,:,1] + delta_u, 0, 255).astype(np.uint8)
    imagen_yuv[:,:,2] = np.clip(imagen_yuv[:,:,2] + delta_v, 0, 255).astype(np.uint8)

    # Paso 4: Transformar la imagen YUV al espacio RGB
    imagen_corregida = cv2.cvtColor(imagen_yuv, cv2.COLOR_YUV2BGR)

    return imagen_corregida, imagen_yuv
```

```python
# Cargar la imagen
imagen = cv2.imread('CONEJO.jpg')

# Aplicar la corrección de balance de blancos y obtener la imagen en el espacio YUV
imagen_corregida, imagen_yuv = balance_de_blancos_yuv(imagen)

# Crear una ventana externa
cv2.namedWindow('Imagenes', cv2.WINDOW_NORMAL)
cv2.resizeWindow('Imagenes', 1200, 400)  # Ajusta el tamaño según sea necesario

# Concatenar las tres imágenes horizontalmente
imagenes_concatenadas = np.concatenate((imagen, imagen_yuv, imagen_corregida), axis=1)

# Agregar títulos a cada imagen
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(imagenes_concatenadas, 'Imagen Original', (10, 30), font, 1, (255, 255, 255), 2, cv2.LINE_AA)
cv2.putText(imagenes_concatenadas, 'Imagen YUV', (imagen.shape[1] + 10, 30), font, 1, (255, 255, 255), 2, cv2.LINE_AA)
cv2.putText(imagenes_concatenadas, 'Imagen Balanceada', (2 * imagen.shape[1] + 10, 30), font, 1, (255, 255, 255), 2, cv2.LINE_AA)

# Mostrar las imágenes en la ventana externa
cv2.imshow('Imagenes', imagenes_concatenadas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
